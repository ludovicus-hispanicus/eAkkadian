# L4.1 Weak Verbs 
A weak verb is a verb whose root consists of only two consonants and one vowel, a biconsonantical base.

1. They can have a long vowel between the two consonants, so called **hollow roots** or **Verbs II-weak**, or after the second radical, so called **final weak verbs** or **Verbs III-weak**:

  ```{admonition} Example
  - **Verbs II weak**: *d창kum* "to kill", *d창rum* "to turn around"
  - **Verbs III weak**: **ban청m* "to build", *qab청* "to say
  ```

2. If the vowel is short the biconsonantical base is expanded by gemination of the second radical by making a triconsonantical by adding a root augment suchas *w(a)* or *n(a)*
