```
# L7.3: N-stem
The main function of this stem is the passive of the G. 

## Meaning

1. **Passive** of the active-transitive G verbs. This is the main function:

    ```{admonition} Example
    - N-stem Pres.: *immahhaá¹£* "he will be smiten"
    - N-stem Perf.: *ittaÅ¡kanÅ«* "They (masc.) have been plaed"
    ```

2. **Middle** voice. 

    ```{admonition} Example
    - G-stem Pret.: *bÄbam ipte* "She opened the gate"
    - N-stem Pret.: *bÄbum ippeti* "The gate opened" (or "the gate was opened")
    ```

3. **Reflexive** and **Reciprocal**. Both can happen, but it is rare. 

    ```{admonition} Example
    - G-stem Inf.: *labÄÅ¡um* "to clothe" â†’ N Perf.: *ittalbaÅ¡* "he clothed himself"
    - G-stem Inf.: *amÄrum* "to see" â†’ N Inf.: *nanmurum* "to see one another, to meet"
    ```

4. Normally state verbs are not attested in N-stem. Nevertheless some of them have N-stem with an **ingressive** meaning, i.e. that something come to existence. 

    ```{admonition} Example
    - G-stem Pres.: *ibaÅ¡Å¡i* "he exists"â†’ N-stem Pres.: *ibbaÅ¡Å¡i* "he comes into existence", "he originates"
    ```

5. Some verbs occur only in N-stem, i.e. the basic form is the N-stem. 

    ```{admonition} Example
    - N-stem Inf.: *napruÅ¡um* "to fly"
    - G-stem Inf.: *palÄsum* "to see" (very rare) â†’ N-stem Inf.: *naplusum* "to see, to look at"
    ```

## Morphological Characteristics

1. The N-stem is characterized  by an *n* before the root.
2. Forms with prefixes the *n* stand before another consonant, thus it assimilates:

   |               | N-Stem      | 
   | -------       | -------------- | 
   | Durative      | *ipparras* <**inparras*
   | Perfect       | *ittapras* <**intapras*
   | Preterite     | *ipparis* <**inparis*
   | Particple     | *mupparsum* <**munparsum*


3. The other form begin with *na*, thus there is no assimilation:

   |               | N-Stem      | 
   | -------       | -------------- | 
   | Imperative    | *napris* 
   | Verbal Adjective | *naprusum*
   | Infinitve     | *naprusum*
   | Stative       | *naprus*


4. Because the N-stem is based on the G-stem the personal prefixes are the same, unlike the Å  and D-stems.
5. The vowel class depends in part of that of the G-stem
   1. Vowel class *a-u*, *a* and *u* â†’ N-stem is *a-i*
   2. Vowel class *i* â†’ N-stem is *i*

   |               | *a-u*-class    | *a*-class      | *u*-class      | *i*-class      | 
   | -------       | -------------- | -------------- | -------------  | -------------- |
   | Durative      | *ipparr**a**s* | *iï¸á¹£á¹£abb**a**t* | *issahh**a**r* | *ippaqq**i**d* |
   | Preterite     | *ippar**i**s*  | *iá¹£á¹£ab**i**t*  | *issah**i**r*  | *ippaq**i**d*  |
   | Infinitve     | *naprusum*     | *naá¹£butum*     | *nashurum*     | *napqudum*     |

    ```{note}
    The *e*-type verbs have *e* instead *a*.
    ```

6. When suffixes are added to the preterite, syncope occurs:
   
   ```{admonition} Example
   - N-stem Pres.: *ipparris* â†’ with ventive: *ipparrsam*
   - N-stem Pres.: *ipparris* â†’ with 3p.pl.: *ipparrsÅ«*
   ``` 

7. The *status constructus* of the participle:  

   ```{admonition} Example
   - N-stem Part. masc.: *mupparsum* â†’ stat. const.: *mupparis*
   - N-stem Part. fem.: *mupparistum* â†’ stat. const.: *mupparsat*
   
8. The infitive, Verbal Adjective and Stative the vowel between *Râ‚‚* and *Râ‚ƒ* is always *u*

   |               | N-Stem      | 
   | -------       | -------------- | 
   | Verbal Adjective | *napr**u**sum*
   | Infinitve     | *napr**u**sum*
   | Stative       | *napr**u**s*

9.  In cuneiform confusion with the G-stem is possible, when the doubling is not indicated in the script:

   ```{admonition} Example
   - <font size="11">ğ’„¿ğ’‰ºğ’† ğ’€‰</font> = *ipaqqid* or *ippaqqid*
   ``` 

## I-*n* Verbs
These verbs present no difficulties in Durative, Preterite or Participle. Nevertheless in the Perfect, the *n* of the root is not assimilated. in the Imperative, Infinitive, Verbal Adjective and Stative is optional:

|               | *naqÄru* (*a-u*)   | *nadÄnu* (*i*) | 
| -------       | -------------- | -------------- |
| Durative      | *innaqqar*     | *innaddin*     |
| Perfect       | *ittanqar*     | *ittandin*     |
| Preterite     | *innaqer*      | *innaddin*     |
| Imperative    | *naqqer*/*nanqer* | *naddin*/*nandin* | 
| Participle    | *munnaqrum*    | *munnadnum*    |
| Verbal Adjective | *naqqurum*/*nanqurum* | *naddunum*/*nandunum*
| Infinitve     | *naqqurum*/*nanqurum*    | *naddunum*/*nandunum*     |
| Stative       | *naqqur*/*nanqur* | *naddun*/*nandun* |

## I-*Ê¾*
In all forms the inital *Ê¾* is replaced by *n*, which is not assimilated to a following radical. Thus, the forms look like the N-stem of I-*n*. The *e* colored verbs may have the expected theme-vowel (*e-i*) or *u* between *Râ‚‚* and *Râ‚ƒ*.

|               | *ahÄzum* (*a-u*) | *ezÄ“bu* (*i*)  | *epÄ“Å¡um* (*e-u*)  | *enÃ»m* (III-*i*) | 
| -------       | -------------- | ------------ | -------------         | ------------- |
| Durative      | *innahhaz*     | *innezzib*   | *inneppeÅ¡*/*inneppuÅ¡* | *innenni*
| Perfect       | *ittanhaz*     | *ittenzib*   | *ittepeÅ¡*/*ittepuÅ¡*   | *ittenni*
| Preterite     | *innahiz*      | *innezib*    | *innepiÅ¡*/*innepuÅ¡*   | *inneni*
| Imperative    | *nanhiz*       | *nenzib*     | *nenpiÅ¡*/*nenpuÅ¡*     | *nenni*
| Participle    | *munnahzum*    | *munnezbum*  | *munnepÅ¡um*           | *munnenÃ»m*
| Verbal Adjective | *nanhuzum*  | *nenzubum*   | *nenpuÅ¡um*            | *nennÃ»m*
| Infinitve     | *nanhuzum*     | *nenzubum*   | *nenpuÅ¡um*            | *nennÃ»m*
| Stative       | *nanhuz*       | *nenzub*     | *nenpuÅ¡*              | *nennu*

```{note}
Some verbs I-*Ê¾* might preserve a strong conjugation throughout the N-paradigm. 
- *adÄrum* G (a-u) "to fear", *naÊ¾durum* 'to be feared": Durative *iÊ¾Ê¾addar*, Perfect *ittaÊ¾dar*, Preterite *iÊ¾Ê¾ader*, etc.
```
